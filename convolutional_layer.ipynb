convolution performed between an image I and a filter(kernel) F and get convolved feature as a output.

Filter passes successively through every pixel of the input image. In each step element wise multiplication is performed between pixels of the image and filter values and then sum up the results into a single output (here it is 4)

*1 x 1 + 1 x 0 + 1 x 1  + 0 x 0 + 1 x 1 + 1 x 0 + 0 x 1 + 0 x 0 +1 x 1 = 4*

We see that the output is 3x3 while the input was 5x5. Now lets see how the output shape is determined.

**Equation for calculating output size:**

input : nh x nw x nc

output :  nH x nW x nC

Equations for calculating output size:

                  nH = ( nh + 2 p - f ) / s + 1

                  nW = ( nw + 2 p - f ) / s +1


**Example:**

input:

  * image :  nh x nw x nc  :  5 x 5 x 1

  * filter :    nh x nw x nc  :  3 x 3 x 1

  * here in this example stride is s =1 , padding is p = 0

output :  nH x nW x nC

nH = `(nh + 2 p - f )/s + 1 `

    = (5 + 2 x 0 - 3)/1 + 1 = 2 + 1 = 3

output :3X3


**Note:** Always number of channels in input image and number of channels in filters should be always equal.

Convolutional layer
* this layer is used to extract features from the input image.
* learning happens in this layer
* When the convolutional operator is applied output feature maps size is reduced that is it not same as the input so in order to make the output same as input we apply padding to the input image.




*   Conv1D --> extract local features
          input (batch_size,input_channels, sequnce length)
*  Conv2D --> extract spatial features
          input (batch_size, input_channels,height,width)
*   Conv3D --> extract volume features across all three spatial dimensions
          input (batch_size,input_channels,depth,hegith,width)
* each differ in dimensions and also in parameters

```python
inpt = torch.randn(20,16,50,100)
inpt.shape
```

```python
conv = nn.Conv2d(16,30,3,stride=1)
```

```python
outpt = conv(inpt)
outpt.shape
```
