* The terms "2D" and "3D" refers to the dimensions of the input data in Neural Networks.
* The 2D data can be moslty seen in images
    * In image each point represent the pixel with the dimensions as width and height
    * Pixel is a color or intensity value at that position in Imaged.
    * 2D data mainly used in image classification, object detection, image segmentation and other computer vision tasks
* 3D = 2D + one more dimesion(i.e depth).
This dimension can represent different channels, frames, or slices of the data
    * 3D data used for video sequence, MRI scans and CT scans( medical scans), volumetric data.

**Note:** To convert a MaxPool3d to a MaxPool2d, one need to adjust the pooling window size and stride accordingly.

````python
import torch
import torch.nn as nn
````


**syntax:**
```
nn.MaxPool3d(kernel_size, stride=None, padding=0, dilation=1, return_indices=False, ceil_mode=False)

```


```
nn.MaxPool2d(kernel_size, stride=None, padding=0, dilation=1, return_indices=False, ceil_mode=False)

```
**Maxpool2d:**

The parameters kernel_size, stride, padding, dilation can either be:

a single int – in which case the same value is used for the height and width dimension

a tuple of two ints – in which case, the first int is used for the height dimension, and the second int for the width dimension

**MaxPool3d:**

The parameters kernel_size, stride, padding, dilation can either be:

a single int – in which case the same value is used for the depth, height and width dimension

a tuple of three ints – in which case, the first int is used for the depth dimension, the second int for the height dimension and the third int for the width dimension


image.png

````python
# input data
inpt = torch.rand(10,12,94,24)
````

````python
maxpol3d = nn.MaxPool3d(kernel_size=(1,3,3),stride=(4,1,2))
````

````python
maxpol2d = nn.MaxPool2d(kernel_size=(3,3),stride=(1,1))
````

By applying MaxPool3d with kernel_size=(1, 3, 3) and stride=(4, 1, 2) on the input data inpt, the resulting output has the shape torch.Size([10, 3, 92, 11]).

To convert this operation to 2D, you can apply MaxPool2d with kernel_size=(3, 3) and stride=(1, 1) on the input data inpt. The resulting output will have the shape torch.Size([10, 12, 92, 22]).

To extract a slice of the output from the MaxPool2d operation, you can perform a slice operation along the depth dimension (dimension 4) and the width dimension (dimension 2). These numbers, 4 and 2, are derived from the stride parameter of the MaxPool3d operation.

````python
output_maxpol3d = maxpol3d(inpt)
output_maxpol3d.shape
````

````python
output_maxpol2d = maxpol2d(inpt)
output_maxpol2d.shape
````

````python
output_maxpol2d[:,::4,:,::2].shape
````

````python
# True if two tensors have the same size and elements, False otherwise.
torch.equal(output_maxpol2d[:,::4,:,::2],output_maxpol3d)
````
